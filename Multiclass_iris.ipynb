```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier

#Keras is subset of tensorflow but it can be called independently
```

```python
path = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(path, header = None)
df.head()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.preprocessing import LabelEncoder     #to encode Y
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
x
```

```python
y
```

```python
y = np_utils.to_categorical(y)       #converting y into one hot encoder as the activation function "softmax" used below require such input for multiclass
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state=1)
```

```python
ann = Sequential()

ann.add(Dense(units = 8, activation = "relu"))
ann.add(Dense(units = 4, activation = "relu"))

ann.add(Dense(units = 3, activation = "softmax"))       #Since the output (y) contains 3 categories of data and softmax requires 3 neurons in output layer

ann.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ["accuracy"])   #for multiclass we use categorical_crossentropy

ann.fit(xtrain, ytrain, epochs = 200)
```

```python
ypred = ann.predict(xtest)
```

Since the output is in one hot encoded form and we will get 3 values for a particular ouput(eg. [1,0,0]) which makes calculation of accuracy difficult, hence converting the output to label encoding again before calculating accuracy

```python
length = len(ypred)
```

```python
ylabel = np.argmax(ytest, axis = 1)     #np.max() function calculates max of an array and np.argmax() fuction finds index of max value
ypredlabel = np.argmax(ypred, axis = 1)
```

```python
np.sum(ylabel==ypredlabel)/length     #accucary
```

```python
def mymodel():
  ann = Sequential()

  ann.add(Dense(units=8, activation="relu"))
  ann.add(Dense(units=4, activation="relu"))

  ann.add(Dense(units=3, activation="softmax"))

  ann.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
  return ann
```

```python
esti = KerasClassifier(build_fn=mymodel, epochs=200, batch_size=5, verbose=0)
results = cross_val_score(esti, x,y , cv=10)
print(results.mean())
```

for output layer we use "Sigmoid" activation finction for binary category and this function only predicts chances of 1 category  (i.e. from 0 and 1, it predicts only chances of 1)(thus there is only 1 neuron in output layer) and it "assumes" remaining chances is of other category which is suitable for binary problems.
Now incase of multiclass categorical data, we cannot use sigmoid(it will give accuracy of each category out of 100, eg. for iris it can give 85% for setosa, 30% for versicolor, & 40% forverginica) thus we use "softmax" function which computes accuracy of each category parallely and gives accurate results and thus it requires input in the form of one hot ecoding. Also, as it predicts all the 3 categories, it requires 3 neurons in its output layer
